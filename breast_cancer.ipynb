# Data Preprocessing

```python
# input variable
X = cancer_df.drop(['target'], axis = 1)
X.head(5)
```

```python
y = cancer_df['target']
y.head(5)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 1)
```

# Feature Scaling

 Converting different units and magnitude data in one unit

```python
#Feature Scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train_sc = sc.fit_transform(X_train)
X_test_sc = sc.transform(X_test)
```

# Model Building

```python
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
```

# Support Vector Classifier

```python
from sklearn.svm import SVC
svc_classifier = SVC()
svc_classifier.fit(X_train, y_train)
y_pred_sv = svc_classifier.predict(X_test)
accuracy_score(y_test, y_pred_sv)
```

```python
# Train with standard scaled data
svc_classifier2 = SVC()
svc_classifier2.fit(X_train_sc, y_train)
y_pred_scv = svc_classifier2.predict(X_test_sc)
accuracy_score(y_test, y_pred_scv)
```

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
lr_classifier = LogisticRegression(random_state = 50, penalty = 'l1')
lr_classifier.fit(X_train, y_train)
y_pred = lr_classifier.predict(X_test)
accuracy_score(y_test, y_pred)
```

```python
#Train with standard scale
lr_classifier2 = LogisticRegression(random_state = 50, penalty = 'l1')
lr_classifier2.fit(X_train_sc, y_train)
y_pred_lr = lr_classifier2.predict(X_test_sc)
accuracy_score(y_test, y_pred_lr)
```

# K-Nearest Neighbor Classifier

```python
from sklearn.neighbors import KNeighborsClassifier
knn_classifier = KNeighborsClassifier(n_neighbors = 5, metric = 'minkowski')
knn_classifier.fit(X_train, y_train)
y_pred_knn = knn_classifier.predict(X_test)
accuracy_score(y_test, y_pred_knn)
```

```python
#Train with standard scale
knn_classifier2 = KNeighborsClassifier(n_neighbors = 5, metric = 'minkowski')
knn_classifier2.fit(X_train_sc, y_train)
y_pred_knn = lr_classifier2.predict(X_test_sc)
accuracy_score(y_test, y_pred_knn)
```

# Naive Bayes Classifier

```python
from sklearn.naive_bayes import GaussianNB
nb_classiier = GaussianNB()
nb_classiier.fit(X_train, y_train)
y_pred_nb = nb_classiier.predict(X_test)
accuracy_score(y_test, y_pred_nb)
```

```python
#Train with standard scale
nb_classifier2 = GaussianNB()
nb_classifier2.fit(X_train_sc, y_train)
y_pred_nb = lr_classifier2.predict(X_test_sc)
accuracy_score(y_test, y_pred_nb)
```

# Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier
dtc_classifier = DecisionTreeClassifier(criterion = 'gini', random_state = 20)
dtc_classifier.fit(X_train, y_train)
y_pred_dtc = dtc_classifier.predict(X_test)
accuracy_score(y_pred_dtc, y_test)
```

```python
#Train with standard scale
dtc_classifier = DecisionTreeClassifier(criterion = 'gini', random_state = 20)
dtc_classifier2 = DecisionTreeClassifier()
dtc_classifier2.fit(X_train_sc, y_train)
y_pred_dtc = lr_classifier2.predict(X_test_sc)
accuracy_score(y_test, y_pred_dtc)
```

# Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
rf_classifier = RandomForestClassifier(n_estimators = 10, criterion = 'gini')
rf_classifier.fit(X_train, y_train)
y_pred_rf = rf_classifier.predict(X_test)
accuracy_score(y_test, y_pred_rf)
```

```python
#Train with standard scale
rf_classifier2 = RandomForestClassifier(n_estimators = 10, criterion = 'entropy')
rf_classifier2.fit(X_train_sc, y_train)
y_pred_rf = rf_classifier2.predict(X_test_sc)
accuracy_score(y_test, y_pred_rf)
```

# Adaboost Classifier

```python
from sklearn.ensemble import AdaBoostClassifier
ad_classifier = AdaBoostClassifier(DecisionTreeClassifier(criterion = 'gini', random_state = 1),n_estimators = 2000,learning_rate = 0.1, algorithm ='SAMME.R',random_state=1)
ad_classifier.fit(X_train,y_train)
y_pred_ad = ad_classifier.predict(X_test)
accuracy_score(y_test, y_pred_ad)
```

```python
#Train with standard scale
ad_classifier2 = AdaBoostClassifier(DecisionTreeClassifier(criterion = 'gini', random_state = 1),n_estimators = 2000,learning_rate = 0.1, algorithm ='SAMME.R',random_state=1)
ad_classifier2.fit(X_train_sc, y_train)
y_pred_ad = rf_classifier2.predict(X_test_sc)
accuracy_score(y_test, y_pred_ad)
```

```python
from sklearn.model_selection import cross_val_score
cross_validation = cross_val_score(estimator= dtc_classifier2, X = X_train_sc, y = y_train, cv = 10)
print("Cross validation of Decesion Tree model with standardScaler = ", cross_validation )
print("Cross validation of Decision Tree model(in mean) with SS = ", cross_validation.mean())
cross_validation = cross_val_score(estimator=lr_classifier, X = X_train, y = y_train, cv = 10)
print("cross validation of Decision Tree model = ", cross_validation)
print("cross validation of Decision Tree model(in mean) = ", cross_validation.mean())

```
